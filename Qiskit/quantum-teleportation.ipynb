```python
from qiskit import *
```

```python
from matplotlib import style
style.use("dark_background")
```

```python
circuit = QuantumCircuit(3, 3)
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
circuit.x(0)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(1) # hadamard gate on qubit 1
circuit.cx(1,2) # applying a controlled X gate between q_1 and q_2
```

```python
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output = 'mpl')
```

The above is the quantum teleportation protocol

```python
circuit.measure(2,2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
print(counts)
```

So we have taken the state from q_0 and teleported it to the q_2. Entagling qubits (e.g. q_0 with q_1 and q_1 with q_2) gives us the ability to tranfer quantum information over long distnaces
